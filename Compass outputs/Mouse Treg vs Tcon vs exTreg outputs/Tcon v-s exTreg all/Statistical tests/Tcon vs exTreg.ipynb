```python
import pandas as pd

# Load your file
df = pd.read_csv("final_stats_with_metadata.csv")  # Replace with your actual file path

# Define classification function
def classify(row):
    if row["adjusted_pval"] < 0.05 and row["cohens_d"] > 0:
        return "p<0.05_Pos"
    elif row["adjusted_pval"] < 0.05 and row["cohens_d"] < 0:
        return "p<0.05_Neg"
    elif row["adjusted_pval"] >= 0.05 and row["cohens_d"] > 0:
        return "p≥0.05_Pos"
    elif row["adjusted_pval"] >= 0.05 and row["cohens_d"] < 0:
        return "p≥0.05_Neg"
    else:
        return None  # Exclude zeros or missing effect sizes

# Apply classification
df["category"] = df.apply(classify, axis=1)

# Group and count
summary = df.groupby(["subsystem", "category"]).size().unstack(fill_value=0)

# Ensure all categories are present
for col in ["p<0.05_Pos", "p<0.05_Neg", "p≥0.05_Pos", "p≥0.05_Neg"]:
    if col not in summary.columns:
        summary[col] = 0

# Reorder columns
summary = summary[["p<0.05_Pos", "p<0.05_Neg", "p≥0.05_Pos", "p≥0.05_Neg"]]

# Reset index to make it a regular column
summary = summary.reset_index()

# Save if needed
summary.to_csv("subsystem_significance_summary.csv", index=False)

# Display a few rows
print(summary.head())
```

```python
import pandas as pd

df = pd.read_csv("final_stats_with_metadata.csv")  

df["direction"] = df["Unnamed: 0"].apply(lambda x: "Pos" if "_pos" in x else ("Neg" if "_neg" in x else None))

df = df.dropna(subset=["direction"])

df["category"] = df["adjusted_pval"].apply(lambda p: "p<0.05" if p < 0.05 else "p≥0.05")

df["group"] = df["category"] + "_" + df["direction"]

summary = df.groupby(["subsystem", "group"]).size().unstack(fill_value=0)

for col in ["p<0.05_Pos", "p<0.05_Neg", "p≥0.05_Pos", "p≥0.05_Neg"]:
    if col not in summary.columns:
        summary[col] = 0

summary = summary[["p<0.05_Pos", "p<0.05_Neg", "p≥0.05_Pos", "p≥0.05_Neg"]]

summary.reset_index(inplace=True)

summary.to_csv("subsystem_significance_summary_only_pval.csv", index=False)

print(summary.head())
```

```python
import pandas as pd
from scipy.stats import fisher_exact

# Load your data
df = pd.read_csv("subsystem_significance_summary_only_pval.csv")  # Replace with your file name

# Initialize result columns
pvals = []
odds_ratios = []
labels = []

# Total number of positive and negative samples (adjust as needed)
total_positive = 83
total_negative = 83

# Loop through each row
for i, row in df.iterrows():
    a = row['p<0.05_Pos']
    b = row['p<0.05_Neg']
    
    # Handle missing values
    if pd.notnull(a) and pd.notnull(b):
        c = total_positive - a
        d = total_negative - b

        # Check if all values are valid
        if all(val >= 0 for val in [a, b, c, d]) and (a + b + c + d > 0):
            table = [[a, b], [c, d]]

            try:
                odds_ratio, p_value = fisher_exact(table)
                pvals.append(p_value)
                odds_ratios.append(odds_ratio)

                if p_value < 0.05:
                    label = "p<0.05_Pos" if odds_ratio > 1 else "p<0.05_Neg"
                else:
                    label = "NS"
            except Exception as e:
                pvals.append(None)
                odds_ratios.append(None)
                label = "Error"
        else:
            pvals.append(None)
            odds_ratios.append(None)
            label = "Invalid Table"
    else:
        pvals.append(None)
        odds_ratios.append(None)
        label = "Missing"

    labels.append(label)

# Add results to DataFrame
df['p_value'] = pvals
df['odds_ratio'] = odds_ratios
df['label'] = labels

# Save result
df.to_csv("fisher_test_output.csv", index=False)
df.head()
```

```python
import pandas as pd
from scipy.stats import fisher_exact
from statsmodels.stats.contingency_tables import Table2x2

df = pd.read_csv("subsystem_significance_summary_only_pval.csv")
df = df.dropna(how='all')
totals = df.iloc[-1]

# Extract counts
a = int(totals['p<0.05_Pos'])  # Significant Positive
b = int(totals['p<0.05_Neg'])  # Significant Negative
c = int(totals['p>0.05_Pos'])  # Non-significant Positive
d = int(totals['p>0.05_Neg'])  # Non-significant Negative

table = [[a, b],
         [c, d]]

oddsratio, p_value = fisher_exact(table)

table2x2 = Table2x2(table)
ci_low, ci_upp = table2x2.oddsratio_confint()

print("Contingency Table:", table)
print("Odds Ratio:", oddsratio)
print("95% CI:", (ci_low, ci_upp))
print("P-Value:", p_value)
```

```python
import pandas as pd

# Load your data
df = pd.read_csv("final_stats_with_metadata.csv")

# Function to extract top N rows with unique subsystems
def get_top_unique_subsystems(df, top_n):
    top_df = df.nsmallest(top_n, 'adjusted_pval')
    unique_df = top_df.drop_duplicates(subset='subsystem', keep='first')
    return unique_df[['subsystem', 'associated_genes']]

# Get each group separately
top10 = get_top_unique_subsystems(df, 10)
top25 = get_top_unique_subsystems(df, 25)
top50 = get_top_unique_subsystems(df, 50)
top75 = get_top_unique_subsystems(df, 75)

# Get top 10 frequent subsystems across the entire dataset
top_subsystems = df['subsystem'].value_counts().head(10).index.tolist()
top10_frequent = (
    df[df['subsystem'].isin(top_subsystems)]
    .drop_duplicates(subset='subsystem', keep='first')
    [['subsystem', 'associated_genes']]
)

# Print each section clearly
print("🔹 Top 10 Adjusted P-value Subsystems:\n", top10, "\n")
print("🔹 Top 25 Adjusted P-value Subsystems:\n", top25, "\n")
print("🔹 Top 50 Adjusted P-value Subsystems:\n", top50, "\n")
print("🔹 Top 75 Adjusted P-value Subsystems:\n", top75, "\n")
print("🔹 Top 10 Frequent Subsystems (based on occurrence):\n", top10_frequent, "\n")

# Optional: Save each as separate file
top10.to_csv("top10_subsystems.csv", index=False)
top25.to_csv("top25_subsystems.csv", index=False)
top50.to_csv("top50_subsystems.csv", index=False)
top75.to_csv("top75_subsystems.csv", index=False)
top10_frequent.to_csv("top10_frequent_subsystems.csv", index=False)
```
