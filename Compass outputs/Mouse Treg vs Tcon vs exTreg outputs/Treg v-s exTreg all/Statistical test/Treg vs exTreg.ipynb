```python
import pandas as pd
df = pd.read_csv('final_stats_with_metadata.csv')

unique_pathways = df['reaction_name'].dropna().unique()
num_pathways = len(unique_pathways)

unique_subsystems = df['subsystem'].dropna().unique()
num_subsystems = len(unique_subsystems)

pathway_df = pd.DataFrame({'reaction_name': unique_pathways})
subsystem_df = pd.DataFrame({'subsystem': unique_subsystems})

pathway_count_df = pd.DataFrame({'reaction_name': [f'Total pathways: {num_pathways}']})
subsystem_count_df = pd.DataFrame({'subsystem': [f'Total subsystems: {num_subsystems}']})

final_pathway_df = pd.concat([pathway_count_df, pathway_df], ignore_index=True)
final_subsystem_df = pd.concat([subsystem_count_df, subsystem_df], ignore_index=True)

with pd.ExcelWriter('final_stats_pathways_and_subsystems.xlsx') as writer:
    final_pathway_df.to_excel(writer, sheet_name='Pathways', index=False)
    final_subsystem_df.to_excel(writer, sheet_name='Subsystems', index=False)
```

```python
import pandas as pd
from scipy.stats import fisher_exact
import numpy as np

# Load CSV
df = pd.read_csv("subsystem_significance_summary_only_pval.csv")

# Rename columns
df = df.rename(columns={
    'p ?0.05_Pos': 'p>0.05_Pos',
    'p?0.05_Neg': 'p>0.05_Neg'
})

# Ensure numeric values and fill NaNs or blanks with 0
cols = ['p<0.05_Pos', 'p>0.05_Pos', 'p<0.05_Neg', 'p>0.05_Neg']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

# Function to compute Fisher's Exact Test with Laplace correction
def run_fisher(row, correction=True):
    a = row['p<0.05_Pos']
    b = row['p>0.05_Pos']
    c = row['p<0.05_Neg']
    d = row['p>0.05_Neg']
    
    if correction:
        a += 0.5
        b += 0.5
        c += 0.5
        d += 0.5

    table = [[a, b], [c, d]]

    try:
        oddsratio, pval = fisher_exact(table)
    except:
        oddsratio, pval = np.nan, np.nan

    return pd.Series({'odds_ratio': oddsratio, 'fisher_p': pval})

# Run Fisher test
df[['odds_ratio', 'fisher_p']] = df.apply(run_fisher, axis=1)

# Save and display
df.to_csv("subsystem_fisher_results.csv", index=False)
df.head()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

df = pd.read_csv("subsystem_significance_summary_only_pval.csv")

df = df.rename(columns={
    'p ?0.05_Pos': 'p>0.05_Pos',
    'p?0.05_Neg': 'p>0.05_Neg'
})

cols = ['p<0.05_Pos', 'p>0.05_Pos', 'p<0.05_Neg', 'p>0.05_Neg']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)

def run_chi_square(row):
    a = row['p<0.05_Pos']
    b = row['p>0.05_Pos']
    c = row['p<0.05_Neg']
    d = row['p>0.05_Neg']
    table = [[a, b], [c, d]]
    try:
        chi2, pval, dof, expected = chi2_contingency(table)
    except:
        pval = np.nan
    return pd.Series({'chi_p': pval})

df[['chi_p']] = df.apply(run_chi_square, axis=1)

df.to_csv("subsystem_chi_test_results.csv", index=False)

df.head()
```

```python
import pandas as pd

# Load your data
df = pd.read_csv("final_stats_with_metadata.csv")

# Function to extract top N rows with unique subsystems
def get_top_unique_subsystems(df, top_n):
    top_df = df.nsmallest(top_n, 'adjusted_pval')
    unique_df = top_df.drop_duplicates(subset='subsystem', keep='first')
    return unique_df[['subsystem', 'associated_genes']]

# Get each group separately
top10 = get_top_unique_subsystems(df, 10)
top25 = get_top_unique_subsystems(df, 25)
top50 = get_top_unique_subsystems(df, 50)
top75 = get_top_unique_subsystems(df, 75)

# Get top 10 frequent subsystems across the entire dataset
top_subsystems = df['subsystem'].value_counts().head(10).index.tolist()
top10_frequent = (
    df[df['subsystem'].isin(top_subsystems)]
    .drop_duplicates(subset='subsystem', keep='first')
    [['subsystem', 'associated_genes']]
)

# Print each section clearly
print("ðŸ”¹ Top 10 Adjusted P-value Subsystems:\n", top10, "\n")
print("ðŸ”¹ Top 25 Adjusted P-value Subsystems:\n", top25, "\n")
print("ðŸ”¹ Top 50 Adjusted P-value Subsystems:\n", top50, "\n")
print("ðŸ”¹ Top 75 Adjusted P-value Subsystems:\n", top75, "\n")
print("ðŸ”¹ Top 10 Frequent Subsystems (based on occurrence):\n", top10_frequent, "\n")

# Optional: Save each as separate file
top10.to_csv("top10_subsystems.csv", index=False)
top25.to_csv("top25_subsystems.csv", index=False)
top50.to_csv("top50_subsystems.csv", index=False)
top75.to_csv("top75_subsystems.csv", index=False)
top10_frequent.to_csv("top10_frequent_subsystems.csv", index=False)
```
